```python
import math

# ReLU Activation Function
def relu(x):
    return max(0, x)

# Sigmoid Activation Function
def sigmoid(x):
    return 1 / (1 + math.exp(-x))

# Tanh Activation Function
def tanh(x):
    return (math.exp(2*x) - 1) / (math.exp(2*x) + 1)

# Leaky ReLU Activation Function
def leaky_relu(x):
    return x if x > 0 else 0.01 * x

# Softmax Activation Function (for list of values)
def softmax(x_list):
    exps = [math.exp(i) for i in x_list]
    total = sum(exps)
    return [i / total for i in exps]


# ----------- Test -----------
x = -2.0
print("ReLU:", relu(x))
print("Sigmoid:", sigmoid(x))
print("Tanh:", tanh(x))
print("LeakyReLU:", leaky_relu(x))

x_list = [2.0, 1.0, 0.1]
print("Softmax:", softmax(x_list))
```
