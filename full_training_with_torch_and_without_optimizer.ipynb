```python
import torch

# Parameters
W = torch.randn(3, 1, requires_grad=True)
b = torch.zeros(1, requires_grad=True)

# Data
x = torch.randn(10, 3)
y = torch.randn(10, 1)

lr = 0.01

for epoch in range(20):

    # Forward
    y_pred = x @ W + b

    # Loss (MSE)
    loss = ((y_pred - y) ** 2).mean()

    # Backward
    loss.backward()

    # Update manually
    with torch.no_grad():
        W -= lr * W.grad
        b -= lr * b.grad

    # Zero gradients
    W.grad.zero_()
    b.grad.zero_()

    print(f"Epoch {epoch}: Loss = {loss.item()}")
```
