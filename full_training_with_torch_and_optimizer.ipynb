```python
import torch
import torch.nn as nn
import torch.optim as optim

# Model
model = nn.Sequential(
    nn.Linear(3, 4),
    nn.ReLU(),
    nn.Linear(4, 1)
)

# Loss
loss_fn = nn.MSELoss()

# Optimizer
optimizer = optim.Adam(model.parameters(), lr=0.01)

# Dummy data
x = torch.randn(10, 3)
y = torch.randn(10, 1)

# Training loop
for epoch in range(20):

    # 1. Forward pass
    predictions = model(x)

    # 2. Compute loss
    loss = loss_fn(predictions, y)

    # 3. Zero gradients
    optimizer.zero_grad()

    # 4. Backpropagation
    loss.backward()

    # 5. Update weights
    optimizer.step()

    print(f"Epoch {epoch}: Loss = {loss.item()}")


# Inference
x = torch.tensor([[2.0, 3.0, -1.0]])

model.eval()
with torch.no_grad():
    output = model(x)

print(output)
```
